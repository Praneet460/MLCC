<h3>Working with Wine Review Dataset</h3>

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
wine_review = pd.read_csv("winemag-data_first150k.csv", index_col = 0)
wine_review.head(5)
```

```python
wine_review.shape
```

```python
# Barplot on basis of 'province' i.e state the wine is from
wine_review['province'].value_counts().head(10)
```

```python
wine_review['province'].value_counts().head(10).plot.bar(figsize=(10,5), fontsize=16, title='Rankings Given by Wine Magazine')
sns.despine(bottom=True, left=True) # used to remove the black border
```

```python
wine_review['province'].value_counts().head(10).plot.pie(figsize=(10,10), fontsize=20)
plt.gca().set_aspect('equal')
```

California produces far more wine than any other province of the world

```python
# in terms of relative proportion

ax = (wine_review['province'].value_counts().head(10)/len(wine_review)).plot.bar(figsize=(10,5), fontsize=20)
ax.set_title("Rankings Given by Wine Magazine", fontsize=20) # set the fontsize of title
sns.despine(bottom=True, left=True) # used to remove the black border
```

California produces almost a third of wines reviewed in Wine Magazine!

```python
wine_review['points'].value_counts().sort_index().plot.bar(figsize=(10,5), fontsize=20)
```

```python
wine_review['country'].value_counts().plot.bar(figsize=(15,5), fontsize=15)
```

```python
# Plotting Line Charts
wine_review['points'].value_counts().sort_index().plot.line(figsize=(10,5), color='red', fontsize=20 )
plt.xlabel('points')
```

```python
# Plotting Area Plot
wine_review['points'].value_counts().sort_index().plot.area(figsize=(10,5), fontsize=20)
```

```python
# Plotting the Histogram 
wine_review[wine_review['price'] < 200]['price'].plot.hist(figsize=(10,5), fontsize=20)
```

```python
wine_review['price'].plot.hist(figsize=(10,5), fontsize=20)
```

This is the real reason I excluded the >$200 bottles earlier; some of these vintages are really expensive! And the chart will "grow" to include them, to the detriment of the rest of the data being shown.

```python
#Scatter Plot
wine_review[wine_review['price'] > 100].sample(100).plot.scatter(x='price', y='points', figsize=(10,5), fontsize=20)
```

```python
# Scatter Plot
wine_review[wine_review['price'] < 100].plot.scatter(x = 'price', y = 'points', figsize=(10,5), fontsize=20)
```

```python
# Hexplots
# Hexplots help aleviate overplotting like in scatter plot in case of wide range of unique values. 
wine_review[wine_review['price'] < 100].plot.hexbin(x = 'price', y = 'points' , gridsize = 15, figsize = (10,5), fontsize=20)
```

```python
wine_review[wine_review['price'] > 1500]
```

```python
wine_review['points'].plot.hist(figsize=(10,5), fontsize=20)
```

<h3>Working with The Complete Pokemon Dataset</h3>

```python
pokemon = pd.read_csv('pokemon.csv')
pokemon.head(4)
```

```python
pokemon.shape
```

```python
list(pokemon) # list of features 
```

```python
# BarPlot
pokemon['type1'].value_counts().plot.bar(figsize=(10,5), fontsize=20)
```

```python
pokemon['generation'].value_counts().sort_index().plot.bar(figsize=(10,5), fontsize=20)
```

```python
pokemon['is_legendary'].value_counts().plot.bar(figsize=(10,5), fontsize=20)
```

```python
pokemon['is_legendary'].value_counts().plot.pie(figsize=(10,10), fontsize=20)
plt.gca().set_aspect('equal')
```

```python
pokemon['hp'].value_counts().sort_index().plot.line(figsize=(10,5), fontsize=20, color='red')
```

```python
pokemon['weight_kg'].plot.hist(figsize=(10,5), fontsize=20)
```

```python
pokemon[pokemon['weight_kg'] <200]['weight_kg'].plot.hist(figsize=(10,5), fontsize=20)
```

```python
# Scatter Plot
pokemon.plot.scatter(x ='attack', y='defense', figsize=(10,5), fontsize=20)
```

```python
# HexPlots
pokemon.plot.hexbin(x='attack', y='defense', gridsize=20, figsize=(10,5), fontsize=20)
```

```python
pokemon_stats_legendary = pokemon.groupby(['is_legendary', 'generation']).mean()[['attack', 'defense']]
pokemon_stats_legendary
```

```python
# Stacked Plot
pokemon_stats_legendary.plot.bar(stacked=True, figsize=(10,5), fontsize=20)
```

```python
pokemon_stats_by_generation = pokemon.groupby('generation').mean()[['hp', 'attack', 'defense', 'sp_attack', 'sp_defense', 'speed']]
pokemon_stats_by_generation
```

```python
pokemon_stats_by_generation.plot.line(figsize=(10,5), fontsize=20)
```
